```python
from pyspark.sql import functions as F

#ENTSO-E prices
source_path="/Volumes/h2/raw_entsoe/landing/stream_input/prices"
checkpoint_path="/Volumes/h2/raw_entsoe/landing/_checkpoints/prices_stream"
target_table="h2.silver.stream_prices_demo"
schema_table="h2.raw_entsoe.prices_2020_raw"

# optional
file_format = "csv"          # csv/json/parquet
header = True                # for csv
max_files_per_trigger = 1
available_now = True         # True = one-shot micro-batch simulation
# =========================
# STREAM TEMPLATE
# =========================
schema = spark.table(schema_table).schema
reader = (
    spark.readStream
    .schema(schema)
    .option("maxFilesPerTrigger", max_files_per_trigger)
)
if file_format == "csv":
    reader = reader.option("header", str(header).lower())
    stream_df = reader.csv(source_path)
elif file_format == "json":
    stream_df = reader.json(source_path)
elif file_format == "parquet":
    stream_df = reader.parquet(source_path)
else:
    raise ValueError(f"Unsupported file_format: {file_format}")
stream_df = stream_df.withColumn("stream_ingested_at_utc", F.current_timestamp())

writer = (
    stream_df.writeStream
    .format("delta")
    .outputMode("append")
    .option("checkpointLocation", checkpoint_path)
)
if available_now:
    query = writer.trigger(availableNow=True).toTable(target_table)
else:
    query = writer.trigger(processingTime="30 seconds").toTable(target_table)
query.awaitTermination()
print("Streaming completed:", target_table)
display(
    spark.table(target_table)
    .orderBy(F.col("stream_ingested_at_utc").desc())
    .limit(20)
)
```
